### Linear Regression

```python
# 4.2 Linear Regression
# Sale price target year build and overall quality with degree 3

# Pull the year built and overall quality columns
X_LR = dt[['YearBuilt','OverallQual']]
# Pull the Sale Price as the Y
Y_LR = dt[['SalePrice']]
```

```python
## Split the data for use in the model
X_LRTrain, X_LRTest,Y_LRTrain, Y_LRTest = train_test_split(X_LR, Y_LR, test_size=0.2)
```

```python
# Start the time measurement
import timeit
startTime = timeit.default_timer()
```

```python
# Create linear regression model that normalizes the features
LR = LinearRegression(normalize=True)
LR.fit(X_LRTrain, Y_LRTrain)
```

```python
# Predict the Y
Y_LRPred = LR.predict(X_LRTest)
```

```python
# Output the run time of the linear regression code
elapsed = timeit.default_timer() - startTime
print("Elapsed Time: ", elapsed)
```

```python
#MSE
lrMSE = mean_squared_error(Y_LRTest,Y_LRPred)
print("Mean Squared Error: ", lrMSE)
```

```python
# RMSE
lrRMSE = sqrt(mean_squared_error(Y_LRTest, Y_LRPred))
print("Root Mean Squared Error: ", lrRMSE)
```

### Polynomial Regression

```python
# Start the timer to measure for the polynomial regression
startTimer = timeit.default_timer()
```

```python
# Do a linear regression using polynomial features with degree of 3
LR2 = LinearRegression(normalize=True)
polyFeat = PolynomialFeatures(degree=3, include_bias=False)
X_LRPoly = polyFeat.fit_transform(X_LRTrain)
```

```python
# Fit the model
LR2.fit(X_LRPoly, Y_LRTrain)
```

```python
# Transform the test dataset so we can use it for predictions
X_LRTestPoly= polyFeat.fit_transform(X_LRTest)
# Make the prediction
LRpred=LR2.predict(X_LRTestPoly)
```

```python
# Show the elapsed time
elapsed = timeit.default_timer()- startTime
print("Elapsed time: ", elapsed)
```

```python
# MSE
prMSE = mean_squared_error(Y_LRTest, LRpred)
print("Mean Squared Error: ", prMSE)
```

```python
# RMSE
prRMSE = sqrt(mean_squared_error(Y_LRTest, LRpred))
print("Root Mean Squared Error: ", prRMSE)
```
