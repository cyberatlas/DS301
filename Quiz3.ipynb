```
%tensorflow_version 2.x
import tensorflow as tf
```

```
from tensorflow import keras
```

```
import numpy as np
```

```
# Single layer
# units = 1 single neuron 
# input_shape of 1 input
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```
# Optimizer sgd enhances each guess, calculates loss using mean squared error
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```
# Input the given values
xs = np.array([-2, -1, 0, 1, 2, 3, 4, 5, 6, 7,8])
ys = np.array([-11, -8,-5, -2, 1, 4, 7, 10, 13, 16, 19])
```

```
# Fit the model and repeat 500 times
model.fit(xs, ys, epochs=500)
```

```
# Predict the value of 15
print(model.predict([15]))
```

```
# Predict the value of 157
print(model.predict([157]))
```
